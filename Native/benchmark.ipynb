```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 5
threads = np.zeros(n)
walltime = np.zeros(n)
flops = np.zeros(n)
threads = [30,60,120,180,240]
walltime = [269,126,124,188,194]
flops = [64.2,137.1,139.4,91.9,89.1]
```

```python
plt.plot(threads,walltime,"rv",markersize=12)
plt.xlabel("Threads",fontsize=18)
plt.ylabel("Walltime (seconds)",fontsize=18)
plt.ylim([100,300])
```

```python
plt.plot(threads,flops,"rv",markersize=12)
plt.xlabel("Threads",fontsize=18)
plt.ylabel("GFLOP/s",fontsize=18)
plt.ylim([50,160])
```
